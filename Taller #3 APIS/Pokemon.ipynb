## POKEMONS :D

```python
import requests
import pandas as pd
```

```python
url = "https://pokeapi.co/api/v2/pokemon/pikachu"
response = requests.get(url)

print(response.status_code)  # debe ser 200
print(response.json())
```

```python
pokemons = []

for i in range(1, 200):  # primeros 200 Pokémon
    url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    response = requests.get(url) # hacer la solicitud
    
    if response.status_code == 200:
        data = response.json()
        
        # extraer stats
        stats = {stat['stat']['name']: stat['base_stat'] for stat in data['stats']} # crear un diccionario de stats
        
        pokemons.append({ # crear un diccionario con la información relevante
            "name": data["name"],
            "type": data["types"][0]["type"]["name"],
            "height": data["height"],
            "weight": data["weight"],
            "attack": stats["attack"],
            "defense": stats["defense"],
            "speed": stats["speed"],
            "hp": stats["hp"]
        })
        
```

```python
Hola_profe = pd.DataFrame(pokemons) # convertir la lista de diccionarios a un DataFrame

```
