# Cargar y explorar el dataset

```python
import pandas as pd
data_estudiantes = pd.read_csv("Student_performance_data _.csv")
data_estudiantes.head(10)
```

```python
data_estudiantes.info() #para ver las columnas y filas hay
```

```python
data_estudiantes.describe() # para obtener estadísticas descriptivas de las columnas numéricas del DataFrame
```

# Limpieza de datos

```python
data_estudiantes.isnull().sum() # para verificar si hay valores nulos en el DataFrame
```

```python
data_estudiantes.drop_duplicates(inplace=True) # para eliminar filas duplicadas en el DataFrame, so es que hay alguna
```

```python
data_estudiantes_COPY = data_estudiantes.copy()
data_estudiantes_COPY['Gender'] = data_estudiantes["Gender"].map({0: "Male", 1: "Female"}) #para convertir los numeros en generos y asi trabajar mas facilmente con ellos
data_estudiantes_COPY['Sports'] = data_estudiantes["Sports"].map({0: "No", 1: "Yes"})
data_estudiantes_COPY['Music'] = data_estudiantes["Music"].map({0: "No", 1: "Yes"}) 


data_estudiantes_COPY.head(10)
```

# Filtros

```python
# mirar si las horas de estudio semanal afectan el rendimiento academico, para esto se puede hacer un filtro y luego sacar la media de cada grupo
estudiantes_con_estudio_largo = data_estudiantes_COPY[data_estudiantes_COPY["StudyTimeWeekly"] > 10]
estudiantes_con_estudio_largo["GPA"].mean() # para sacar la media de las notas finales de los estudiantes con estudio largo
print(estudiantes_con_estudio_largo["GPA"].mean())
#aca el promedio de los que estudian mas de 10h a la semana es de 2.0
```

```python
estudiantes_con_estudio_corto = data_estudiantes_COPY[data_estudiantes_COPY["StudyTimeWeekly"] < 10]
estudiantes_con_estudio_corto["GPA"].mean() # para sacar la media de las notas finales de los estudiantes con estudio corto
print(estudiantes_con_estudio_corto["GPA"].mean())
#aca el promedio de los que estudian menos de 10h a la semana es de 1,7
```

 No hay casi diferencia entre ambos pero los que estudian mas tienden a tener un poco mas de nota

```python
#mirar si las actividades extracurriculares y deportes o musica afectan el rendimiento academico
estudiantes_con_deportes_o_musica = data_estudiantes_COPY[
	(data_estudiantes_COPY["Sports"] == "Yes") |
	(data_estudiantes_COPY["Music"] == "Yes")
]
estudiantes_con_deportes_o_musica["GPA"].mean() # para sacar la media de las notas finales de los estudiantes que hacen deportes o música
print(estudiantes_con_deportes_o_musica["GPA"].mean())
```

```python
estudiantes_sin_deportes_ni_musica = data_estudiantes_COPY[
	(data_estudiantes_COPY["Sports"] == "No") &
	(data_estudiantes_COPY["Music"] == "No")
]
estudiantes_sin_deportes_ni_musica["GPA"].mean() # para sacar la media de las notas finales de los estudiantes que no hacen deportes ni música
print(estudiantes_sin_deportes_ni_musica["GPA"].mean())
```

```python
#ver como afecta la absences a clase al rendimiento academico
estudiantes_con_absences_bajas = data_estudiantes_COPY[data_estudiantes_COPY["ClassAttendance"] < 15] # <15 por que es la mitad del maximo de absences que es 29
estudiantes_con_absences_bajas["GPA"].mean() # para sacar la media de las notas finales de los estudiantes con absences bajas
print(estudiantes_con_absences_bajas["GPA"].mean())
```

Mi hipotesis a partir de lo visto es la siguiente:

Los estudiantes que tienden a ser estudiar mas tienen mejores notas

Los estudiantes que realizan alguna actividad recreativa tienden a tener mejores notas
