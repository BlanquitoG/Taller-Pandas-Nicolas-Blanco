Preguntas a resolver:


1 ¿Cuántos sobrevivieron?

2 ¿Qué porcentaje de pasajeros sobrevivió?

3 ¿Sobrevivieron más hombres o mujeres?

4 ¿Qué clase social (Pclass) tuvo mayor tasa de supervivencia?

5 ¿Cuál fue la edad promedio de los pasajeros?

6 ¿Los pasajeros de primera clase sobrevivieron más que los de tercera?

7 ¿Las personas jóvenes sobrevivieron más que las mayores?

```python
# 1. ¿Cuántos pasajeros sobrevivieron y cuántos no?
survival_counts = data['Survived'].value_counts()
print(survival_counts) 
```

```python
print("No sobrevivieron:", survival_counts[0])
print("Sobrevivieron:", survival_counts[1])
```

```python
# 2. ¿Cuál es la edad promedio de los pasajeros?
average_age = data['Age'].mean() # el .mean() calcula la media de la columna 'Age'
print(f'La edad promedio de los pasajeros es: {average_age} años')
```

```python
#3. ¿Cuál es la proporción de hombres y mujeres a bordo del Titanic?
gender_counts = data['Sex'].value_counts() #value_counts() cuenta la cantidad de cada valor único en la columna 'Sex'
total_passengers = len(data) #len(() calcula el numero total de filas
proportion_men = gender_counts['male'] / total_passengers # Calcula la proporción de hombres dividiendo el conteo de hombres por el total de pasajeros
proportion_woman = gender_counts['female'] / total_passengers 
print(f'Proporción de hombres: {proportion_men:.2f}') # : .2f es para hacer que solo aparezcan dos decimmales
print(f'Proporción de mujeres: {proportion_woman:.2f}') 
```

```python
# 4. ¿Cuál es la tasa de supervivencia por clase de pasajero?
survival_by_class = data.groupby('Pclass')['Survived'].mean() # Agrupa por clase y calcula la media de supervivencia
print(f'Tasa de supervivencia por clase: {survival_by_class*100}%')
```

```python
# 5. ¿Cuál es la edad promedio de los pasajeros que sobrevivieron y los que no sobrevivieron?
average_edad_sobrevivieron = data[data['Survived'] == 1]['Age'].mean() # Filtra los pasajeros que sobrevivieron y calcula la edad promedio
average_edad_no_sobrevivieron = data[data['Survived'] == 0]['Age'].mean() # Filtra los pasajeros que no sobrevivieron y calcula la edad promedio
print(f'Edad promedio de los pasajeros que sobrevivieron: {average_edad_sobrevivieron:.2f}')
print(f'Edad promedio de los pasajeros que no sobrevivieron: {average_edad_no_sobrevivieron:.2f}')
```

```python
#6. ¿Cuál es la tarifa promedio pagada por los pasajeros que sobrevivieron y los que no sobrevivieron?
average_si_sobrevivieron = data[data['Survived'] == 1]['Fare'].mean() # Filtra los pasajeros que sobrevivieron y calcula la tarifa promedio
average_no_sobrevivieron = data[data['Survived'] == 0]['Fare'].mean() # Filtra los pasajeros que no sobrevivieron y calcula la tarifa promedio
print(f'Tarifa promedio pagada por los pasajeros que sobrevivieron: {average_si_sobrevivieron:.2f}')
print(f'Tarifa promedio pagada por los pasajeros que no sobrevivieron: {average_no_sobrevivieron:.2f}')
```

```python
#7. ¿Cuál es la proporción de pasajeros que viajaban solos (sin familiares) y los que viajaban con familiares?
solo_counts = data[data['SibSp'] == 0]['Survived'].count() 
con_familia_counts = data[data['SibSp'] > 0]['Survived'].count()
total_passengers = len(data)
proportion_solo = solo_counts / total_passengers
proportion_con_familia = con_familia_counts / total_passengers
print(f'Proporción de pasajeros que viajaban solos: {proportion_solo:.2f}')
print(f'Proporción de pasajeros que viajaban con familiares: {proportion_con_familia:.2f}')
```
