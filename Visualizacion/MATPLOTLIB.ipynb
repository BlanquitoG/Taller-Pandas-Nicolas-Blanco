```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 9]

plt.plot(x, y)
plt.title("Evolución de ventas")
plt.xlabel("Mes") # Agrega una etiqueta al eje x
plt.ylabel("Ventas (miles)") # Agrega una etiqueta al eje y
plt.grid(True) # Agrega una cuadrícula al gráfico
plt.show()
```

```python
plt.plot(x, y, label='Producto A')
plt.plot(x, [8, 10, 6, 12, 11], label='Producto B')
plt.title("Comparativo de ventas")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.legend() # Agrega una leyenda para identificar las líneas
plt.show()
```

```python
categorias = ['Enero', 'Febrero', 'Marzo']
valores = [100, 120, 90]

plt.bar(categorias, valores, color='lightblue') 
plt.title("Ventas trimestrales")
plt.ylabel("Miles de pesos")
plt.show()
```

```python
import numpy as np

datos = np.random.normal(0, 1, 1000) # Genera 1000 datos aleatorios con distribución normal
plt.hist(datos, bins=30, color='gray') #bins es el número de barras en el histograma
plt.title("Distribución de datos")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4)) # Crea una figura con 1 fila y 2 columnas de subgráficos

axs[0].bar(['A', 'B', 'C'], [10, 15, 7]) # Gráfico de barras en el primer subgráfico
axs[0].set_title("Gráfico de barras") # Agrega un título al primer subgráfico

axs[1].plot([1, 2, 3], [5, 8, 6]) # Gráfico de líneas en el segundo subgráfico
axs[1].set_title("Gráfico de líneas") # Agrega un título al segundo subgráfico

plt.tight_layout() # Ajusta el diseño para evitar solapamientos
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1️⃣ Preparar datos
meses = ['Enero', 'Febrero', 'Marzo']
ventas_A = [120, 135, 150]
ventas_B = [100, 140, 130]

# -----------------------------------
# 2️⃣ Gráfico de barras Producto A
# -----------------------------------
plt.figure(figsize=(6,4))
plt.bar(meses, ventas_A, color='skyblue')

# Línea de meta
plt.axhline(y=125, color='red', linestyle='--', label='Meta (125)')

plt.title("Ventas Producto A")
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.legend()
plt.show()

# -----------------------------------
# 3️⃣ Gráfico de barras comparativo
# -----------------------------------
x = np.arange(len(meses))
ancho = 0.35

plt.figure(figsize=(6,4))
plt.bar(x - ancho/2, ventas_A, width=ancho, label='Producto A', color='blue')
plt.bar(x + ancho/2, ventas_B, width=ancho, label='Producto B', color='orange')

plt.xticks(x, meses)
plt.title("Comparación de Ventas")
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.legend()
plt.show()

# -----------------------------------
# 4️⃣ Gráfico de líneas
# -----------------------------------
plt.figure(figsize=(6,4))

plt.plot(meses, ventas_A, marker='^', linestyle='--', label='Producto A')
plt.plot(meses, ventas_B, marker='*', linestyle='-.', label='Producto B')

plt.title("Evolución de Ventas")
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.legend()
plt.grid(True)

plt.show()

# -----------------------------------
# 5️⃣ Subplots (opcional)
# -----------------------------------
fig, axs = plt.subplots(1, 2, figsize=(10,4))

# Barras Producto A
axs[0].bar(meses, ventas_A, color='green')
axs[0].axhline(y=125, color='red', linestyle='--', label='Meta')
axs[0].set_title("Producto A")
axs[0].legend()

# Línea Producto B
axs[1].plot(meses, ventas_B, marker='o', linestyle='-.', color='purple', label='Producto B')
axs[1].set_title("Producto B")
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```
